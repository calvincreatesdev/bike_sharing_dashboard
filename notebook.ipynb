## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

### Cleaning Data

#### Fixing redundant problem

```python
# Drop the 'instant' column as it only serves as an index and is redundant
df_day.drop(columns=['instant'], inplace=True, errors='ignore')
df_hour.drop(columns=['instant'], inplace=True, errors='ignore')

# Final check after drop the 'instant' column
display(df_day.head())
```

**Insight:** (Opsional)
- Kondisi Awal (Before): Pada struktur awal dataset, terdapat kolom bernama instant yang nilainya hanya berupa nomor urut baris (1, 2, 3, dan seterusnya). Kolom ini diklasifikasikan sebagai data yang redundan karena struktur DataFrame pada Pandas sudah secara otomatis memiliki indeks baris bawaan.

- Proses Eksekusi (Processing): Pembersihan dilakukan menggunakan metode .drop(columns=['instant']) dengan parameter inplace=True agar penghapusan langsung diterapkan pada dataframe utama. Tambahan parameter errors='ignore' digunakan sebagai langkah preventif untuk memastikan sistem tidak memunculkan error apabila cell kode dieksekusi lebih dari satu kali.

- Hasil Akhir (Output): Kolom instant berhasil dieliminasi dari kedua dataset (df_day dan df_hour). Hasilnya dataset menjadi lebih ringkas.

#### Fixing invalid data type problem

```python
# Convert 'dteday' column from object (string) to datetime format
df_day['dteday'] = pd.to_datetime(df_day['dteday'])
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])

# Final check after converting
display(df_day.head())
```

**Insight:** (Opsional)
- Kondisi Awal (Before): Berdasarkan hasil inspeksi awal, kolom dteday yang merepresentasikan tanggal (contoh: "2011-01-01") terbaca oleh sistem dengan tipe data object atau string (teks biasa). Dalam kondisi ini, sistem tidak mengenali data tersebut sebagai sebuah entitas waktu, sehingga tidak bisa digunakan untuk analisis time-series.

- Proses Eksekusi (Processing): Transformasi dilakukan menggunakan fungsi pd.to_datetime(). Fungsi ini secara spesifik bertugas membaca deretan teks format tanggal dan mengonversinya ke dalam format waktu.

- Hasil Akhir (Output): Tipe data pada kolom dteday berubah dari object menjadi datetime64[ns]. Perubahan fundamental ini memberikan dimensi waktu yang nyata pada dataset, yang nantinya memungkinkan untuk melakukan ekstraksi komponen waktu spesifik, pengurutan temporal, dan agregasi data berbasis kalender pada tahap eksplorasi.

#### Fixing inconsistent value (mapping) problem

```python
# Create mapping dictionaries based on Readme.txt
season_mapping = {1: 'Springer', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
year_mapping = {0: '2011', 1: '2012'}
weather_mapping = {
    1: 'Clear/Partly Cloudy',
    2: 'Mist/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Snow'
}

# Mapping numeric categorical values into human-readable text labels
df_day['season'] = df_day['season'].replace(season_mapping)
df_day['yr'] = df_day['yr'].replace(year_mapping)
df_day['weathersit'] = df_day['weathersit'].replace(weather_mapping)

df_hour['season'] = df_hour['season'].replace(season_mapping)
df_hour['yr'] = df_hour['yr'].replace(year_mapping)
df_hour['weathersit'] = df_hour['weathersit'].replace(weather_mapping)

# Final check after caterogical mapping
display(df_day.head())
```

**Insight:** (Opsional)
- Kondisi Awal (Before): Beberapa variabel kategorikal seperti season (musim), yr (tahun), dan weathersit (situasi cuaca) direpresentasikan menggunakan angka murni (integer). Contohnya, musim dinotasikan dengan angka 1 hingga 4. Format ini tidak konsisten dengan keterbacaan manusia (human-readable) dan sangat berisiko menimbulkan misinterpretasi saat ditampilkan dalam bentuk visualisasi atau dashboard.

- Proses Eksekusi (Processing): Proses dimulai dengan menyusun dictionary yang berpedoman pada dokumen Readme.txt. Setelah dictionary sudah disusun, fungsi .replace() diaplikasikan untuk menukar nilai angka menjadi teks deskriptif secara massal. Saya memilih .replace() karena untuk mencegah timbulnya nilai kosong (NaN) jika terjadi duplikasi eksekusi kode.

- Hasil Akhir (Output): Nilai numerik pada kolom-kolom kategorikal telah berhasil digantikan dengan label teks yang deskriptif dan intuitif (contoh: angka 1 pada season kini terdata sebagai "Springer"). Hal ini memastikan integritas semantik data tetap terjaga dan visualisasi akhir dapat dipahami secara langsung tanpa perlu proses penerjemahan manual.

## Exploratory Data Analysis (EDA)

### Explore df_hour : Pola Jam Sibuk (Workingday vs Holiday)

```python
# Aggregate the average hourly rentals for casual and registered users based on working days
hourly_pattern = df_hour.groupby(by=["workingday", "hr"]).agg({
    "casual": "mean",
    "registered": "mean",
    "cnt": "mean"
}).reset_index()

print("----- Pola Rata-rata Penyewaan per Jam (Berdasarkan Hari Kerja) -----")
display(hourly_pattern.head(48)) # Displaying the first 48 rows (first 24 hours is holiday (0), the second 24 hours is workingday (1))
```

**Insight:** (Opsional)
- **Pola Aktivitas pada Hari Libur (workingday = 0):** Berdasarkan distribusi data pada 24 baris pertama, terlihat bahwa pada akhir pekan atau hari libur, metrik pengguna biasa (casual) menunjukkan tingkat aktivitas yang signifikan. Tren penyewaan mulai mengalami eskalasi pada pukul 10.00 pagi dan mencapai titik puncak (peak hour) secara beruntun pada rentang waktu 12.00 hingga 15.00 siang. Pola lonjakan tunggal di pertengahan hari ini mengindikasikan bahwa pengoperasian sepeda didominasi oleh motivasi rekreasi, wisata, maupun aktivitas luar ruangan yang santai.

- **Pola Aktivitas pada Hari Kerja (workingday = 1):** Terjadi pergeseran karakteristik pengguna yang sangat kontras pada hari kerja. Angka pengguna casual menurun drastis, sementara dominasi penyewaan beralih secara mutlak kepada pengguna terdaftar (registered). Distribusi data pada jam kerja membentuk pola bimodal, dua puncak lonjakan tajam yang terjadi secara spesifik pada pukul 08.00 pagi dan rentang pukul 17.00 hingga 18.00 sore. Pola ini memvalidasi kesimpulan bahwa pelanggan terdaftar memanfaatkan layanan sepeda sebagai sarana transportasi komuter harian untuk menunjang mobilitas keberangkatan dan kepulangan kerja atau kegiatan akademik.

### Explore df_day : Performa Musim, Cuaca dan Tren Bulanan

```python
# Explore the impact of season and weather on average daily rentals
season_weather_impact = df_day.groupby(by=["season", "weathersit"], observed=True).agg({
    "cnt": "mean"
}).reset_index().sort_values(by="cnt", ascending=False)

print("----- Rata-rata Penyewaan Berdasarkan Musim & Cuaca -----")
display(season_weather_impact)

# Explore Month-over-Month (MoM) total rentals for 2011 and 2012
monthly_trend = df_day.groupby(by=["yr", "mnth"]).agg({
    "cnt": "sum"
}).reset_index()

print("\n----- Total Penyewaan Bulanan (2011 vs 2012) -----")
display(monthly_trend.head(12)) # Previewing 2011 data
```

**Insight:** (Opsional)
- **Dampak Signifikan Musim dan Kondisi Cuaca:** Berdasarkan hasil agregasi performa lingkungan, teridentifikasi bahwa kombinasi musim gugur (Fall) dengan cuaca cerah/berawan sebagian (Clear/Partly Cloudy) merupakan kondisi paling optimal yang menghasilkan rata-rata penyewaan harian tertinggi (mencapai 5.878 penyewaan). Sebaliknya, kombinasi musim semi (Springer) dengan presipitasi seperti hujan/salju ringan (Light Snow/Rain) menekan angka penyewaan ke titik terendah. Pola hierarki ini secara konsisten menunjukkan bahwa visibilitas cuaca yang baik dan suhu yang moderat adalah faktor pendorong utama (katalisator) dalam menstimulasi minat bersepeda pelanggan.

- **Pola Fluktuasi Musiman (Seasonality) Bulanan:** Distribusi total penyewaan bulanan pada tahun observasi pertama (2011) memperlihatkan kurva fluktuasi yang selaras dengan siklus iklim tahunan. Volume transaksi berada pada level fundamental (terendah) di awal tahun (Januari - Februari) saat suhu masih dingin, kemudian mengalami eskalasi progresif yang sejalan dengan transisi musim, dan pada akhirnya mencapai titik puncak (peak season ) pada periode pertengahan tahun (Juni - Agustus) dengan mencatatkan lebih dari 140.000 penyewaan per bulan. Hal ini menegaskan bahwa operasional bisnis sangat rentan terhadap dinamika perubahan bulan dan iklim.

### Explore df_day : Faktor Lingkungan (Suhu dan Angin) serta Perilaku Mingguan

```python
# Explore which day of the week brings the highest casual users
weekday_casual = df_day.groupby(by="weekday").agg({
    "casual": "mean",
    "registered": "mean"
}).reset_index().sort_values(by="casual", ascending=False)

print("----- Rata-rata Penyewaan Berdasarkan Hari dalam Seminggu -----")
display(weekday_casual)

# Explore the correlation between environmental factors and bike rentals
env_factors = ["temp", "atemp", "hum", "windspeed", "casual", "registered", "cnt"]
env_correlation = df_day[env_factors].corr()

print("\n----- Matriks Korelasi Faktor Lingkungan -----")
display(env_correlation)
```

**Insight:** (Opsional)
- **Validasi Pola Rekreasi Akhir Pekan:** Melalui agregasi berdasarkan hari (`weekday`), terkonfirmasi bahwa akhir pekan (Sabtu dan Minggu) merupakan titik puncak akumulasi bagi pengguna biasa (casual), dengan rata-rata penyewaan melampaui 1.300 hingga 1.400 transaksi per hari. Dominasi pengguna kasual di akhir pekan ini berbanding terbalik dengan hari kerja (Senin-Jumat), di mana volume transaksi diambil alih secara mayoritas oleh pengguna terdaftar (registered) dengan rata-rata di atas 3.600 penyewaan. Hal ini memvalidasi segmentasi pasar yang terbagi jelas antara kebutuhan rekreasi (akhir pekan) dan utilitas komuter (hari kerja).

- **Pengaruh Signifikan Faktor Lingkungan Terhadap Konversi:** Analisis matriks korelasi memberikan landasan kuantitatif yang membuktikan bahwa fluktuasi lingkungan berdampak langsung pada keputusan pelanggan. Variabel suhu udara (`temp` dan `atemp`) mencatatkan korelasi positif yang substansial (sekitar 0.63) terhadap total penyewaan, mengindikasikan bahwa cuaca hangat menstimulasi peningkatan transaksi. Sebaliknya, kecepatan angin (`windspeed`) dan kelembaban (`hum`) berkorelasi negatif, yang bermakna bahwa elemen cuaca ekstrem tersebut bertindak sebagai friksi operasional yang secara rasional menurunkan minat fisik pelanggan untuk bersepeda.

## Visualization & Explanatory Analysis

###Pertanyaan 1: Bagaimana perbedaan pola jam sibuk (*peak hours*) penyewaan sepeda antara pengguna biasa (*casual*) dan pengguna terdaftar (*registered*) pada hari kerja (*workingday*) dibandingkan hari libur (*holiday*) selama periode tahun 2011-2012?

```python
# Separate working day and holiday data from 'hourly_pattern' EDA output
df_workingday = hourly_pattern[hourly_pattern['workingday'] == 1]
df_holiday = hourly_pattern[hourly_pattern['workingday'] == 0]

# Create a figure with 1 row and 2 columns for subplots
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# --- Subplot 1: Working Day ---
sns.lineplot(data=df_workingday, x='hr', y='registered', label='Registered (Member)', marker='o', color='#1f77b4', ax=ax[0])
sns.lineplot(data=df_workingday, x='hr', y='casual', label='Casual (Biasa)', marker='o', color='#ff7f0e', ax=ax[0])
ax[0].set_title('Pola Jam Sibuk pada Hari Kerja (Workingday)', fontsize=14, pad=15)
ax[0].set_xlabel('Jam (0-23)', fontsize=12)
ax[0].set_ylabel('Rata-rata Penyewaan', fontsize=12)
ax[0].set_xticks(range(0, 24))
ax[0].grid(True, linestyle='--', alpha=0.6)
ax[0].legend()

# --- Subplot 2: Holiday ---
sns.lineplot(data=df_holiday, x='hr', y='registered', label='Registered (Member)', marker='o', color='#1f77b4', ax=ax[1])
sns.lineplot(data=df_holiday, x='hr', y='casual', label='Casual (Biasa)', marker='o', color='#ff7f0e', ax=ax[1])
ax[1].set_title('Pola Jam Sibuk pada Hari Libur (Holiday)', fontsize=14, pad=15)
ax[1].set_xlabel('Jam (0-23)', fontsize=12)
ax[1].set_ylabel('Rata-rata Penyewaan', fontsize=12)
ax[1].set_xticks(range(0, 24))
ax[1].grid(True, linestyle='--', alpha=0.6)
ax[1].legend()

# Add main title and adjust layout
plt.suptitle('Perbandingan Pola Jam Sibuk Penyewaan Sepeda (Casual vs Registered)', fontsize=18, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)
- **Hari Kerja (Grafik Kiri):** Pengguna terdaftar (Registered) mendominasi secara mutlak dengan membentuk pola bimodal (dua puncak). Puncak aktivitas (peak hours) terjadi pada pukul 08.00 pagi dan rentang pukul 17.00 - 18.00 sore. Hal ini membuktikan bahwa sepeda utamanya digunakan sebagai moda transportasi komuter harian. Di sisi lain, pengguna kasual sangat sepi sepanjang hari.

- **Hari Libur (Grafik Kanan):** Pola bergeser drastis di mana penyewaan oleh pengguna Casual meningkat tajam membentuk kurva lonjakan tunggal di pertengahan hari (pukul 12.00 hingga 15.00 siang). Ini mengonfirmasi bahwa sepeda disewa untuk tujuan rekreasi dan olahraga santai pada akhir pekan, bukan untuk mengejar waktu pagi buta.

### Pertanyaan 2: Seberapa besar dampak perubahan kondisi cuaca (*weathersit*) dan musim (*season*) terhadap persentase penurunan rata-rata total penyewaan sepeda harian (*cnt*) sepanjang tahun 2011-2012?

```python
# Create a figure canvas with custom dimensions
plt.figure(figsize=(14, 7))

# Create a grouped bar chart using seaborn
# Use the 'season_weather_impact' dataframe generated from EDA
ax = sns.barplot(
    data=season_weather_impact,
    x='season',
    y='cnt',
    hue='weathersit',
    palette='viridis'
)

# Add title and axis labels
plt.title('Dampak Musim dan Kondisi Cuaca terhadap Rata-rata Penyewaan Harian', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Kategori Musim', fontsize=12, labelpad=10)
plt.ylabel('Rata-rata Penyewaan Harian', fontsize=12, labelpad=10)

# Move the legend outside the plot area so it doesn't block the bars
plt.legend(title='Kondisi Cuaca', bbox_to_anchor=(1.01, 1), loc='upper left')

# Add grid lines behind the bars for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display the exact value on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3, fontsize=10)

# Adjust layout to prevent clipping and render the plot
plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)
- **Titik Puncak (Peak Performance):** Rata-rata penyewaan tertinggi terjadi pada musim gugur (Fall) saat cuaca cerah (Clear/Partly Cloudy), mencapai angka **5.878** penyewaan per hari. Suhu yang sejuk dipadu langit cerah menjadi kondisi paling ideal bagi pesepeda.

- **Titik Terendah (Lowest Performance):** Rata-rata penyewaan terendah tercatat pada musim semi (Springer) ketika terjadi hujan/salju ringan (Light Snow/Rain), yang hanya menyentuh angka **935** penyewaan per hari.

- **Persentase Penurunan Ekstrem:** Terjadi penurunan drastis sekitar **84%** dari titik performa terbaik (5.878) ke titik terburuknya (935). ### Pertanyaan 6: Pada level kecepatan angin (*windspeed*) berapakah rata-rata penyewaan sepeda harian mulai mengalami penurunan drastis, dan apakah dampaknya lebih signifikan terhadap pengguna kasual dibandingkan pengguna terdaftar?

```python
# Create a figure canvas
plt.figure(figsize=(12, 6))

# Define windspeed bins to make the trend smoother and easier to read
# Round the windspeed to 1 decimal place to group them
df_day['wind_bin'] = df_day['windspeed'].round(1)
wind_impact = df_day.groupby('wind_bin').agg({'casual': 'mean', 'registered': 'mean'}).reset_index()

# Plotting both user types to compare the impact
sns.lineplot(data=wind_impact, x='wind_bin', y='registered', label='Registered (Member)', marker='s', color='#1f77b4', linewidth=2)
sns.lineplot(data=wind_impact, x='wind_bin', y='casual', label='Casual (Biasa)', marker='o', color='#ff7f0e', linewidth=2)

# Add title and axis labels
plt.title('Dampak Kecepatan Angin terhadap Rata-rata Penyewaan', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Kecepatan Angin (Normalized Windspeed)', fontsize=12)
plt.ylabel('Rata-rata Penyewaan', fontsize=12)

# Add a vertical line to highlight the drop-off point (around 0.3)
plt.axvline(x=0.3, color='red', linestyle='--', alpha=0.7, label='Critical Drop-off Point')

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)

## Analisis Lanjutan (Opsional)

```python
# Create intensity segments (clustering) based on the total daily rentals (cnt)
# We define the thresholds to represent Low, Moderate, and High Intensity
def categorize_intensity(cnt):
    if cnt < 3000:
        return 'Low Circulation'
    elif 3000 <= cnt <= 6000:
        return 'Moderate Usage'
    else:
        return 'High Turnover'

# Apply the function to the daily dataframe
df_day['usage_intensity'] = df_day['cnt'].apply(categorize_intensity)

# Analyze the relationship between intensity and other environmental factors
intensity_analysis = df_day.groupby('usage_intensity').agg({
    'cnt': ['count', 'mean', 'max'],
    'temp': 'mean',
    'hum': 'mean',
    'registered': 'mean',
    'casual': 'mean'
}).reset_index()

# Sort for better presentation
intensity_analysis.columns = ['Intensity_Category', 'Day_Count', 'Avg_Rentals', 'Max_Rentals', 'Avg_Temp', 'Avg_Humidity', 'Avg_Registered', 'Avg_Casual']
intensity_analysis = intensity_analysis.sort_values(by='Avg_Rentals', ascending=False)

print("=== Hasil Segmentasi Intensitas Penggunaan Sepeda ===")
display(intensity_analysis)
```

**Insight:** (Opsional)
- **High Turnover (Intensitas Tinggi):** Cluster ini merupakan periode di mana armada sepeda bekerja paling keras dengan rata-rata di atas 6.000 penyewaan. Menariknya, pada kategori ini, baik pengguna terdaftar maupun kasual sama-sama menunjukkan angka yang tinggi. Secara operasional, hari-hari dalam cluster ini membutuhkan pengecekan servis harian karena tingginya frekuensi penggunaan (turnover).

- **Moderate Usage (Penggunaan Moderat):** Merupakan kondisi normal operasional dengan rata-rata penyewaan di kisaran 4.000 - 5.000. Sebagian besar hari dalam setahun masuk ke dalam kategori ini, yang menunjukkan stabilitas bisnis yang baik.

- **Low Circulation (Sirkulasi Rendah):** Terjadi pada hari-hari dengan rata-rata penyewaan di bawah 3.000. Cluster ini biasanya bertepatan dengan suhu udara yang rendah atau kondisi cuaca yang kurang mendukung. Dari sisi bisnis, periode ini adalah waktu yang paling efisien untuk melakukan deep maintenance atau servis besar pada unit sepeda tanpa mengganggu ketersediaan armada bagi pelanggan.
